```python
from glob import glob

img_list = glob('origin/*.jpg')

db = {}

for i in range(1,len(img_list)+1) :
    img_name = f'{i}_1'
    img_path = f'origin/{i}_1.jpg'
    results = model(img_path)
    
    tags = set()

    for pred in results.pred[0]:
        tag = results.names[int(pred[-1])] # English
#         tag = tags_ko[int(pred[-1])] # Korean
        tag = tag.replace(' ', '')
        tags.add(tag)
        
    db[img_path] = list(tags)

db
```

```python
import pickle

with open('databses.pickle','wb') as f :
    pickle.dump(db,f)
```

```python
import pickle

with open('databses.pickle','rb') as f :
    lists = pickle.load(f)
lists
```
